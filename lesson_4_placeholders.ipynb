## Chapter: Placeholders

```python
import tensorflow as tf
```

```python
x = tf.placeholder("float", None) # we can have any number of values
y = x * 2

with tf.Session() as session:
    result = session.run(y, feed_dict={x: [1, 2, 3]})
    print(result)
```

```python
# The first dimension of the placeholder is None, meaning we can have any number of rows
x = tf.placeholder("float", [None, 3]) # create a 3 by 2 matrix
y = x * 2

with tf.Session() as session:
    x_data = [[1, 2, 3],
              [4, 5, 6],]
    result = session.run(y, feed_dict={x: x_data})
    print(result)
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os

# First, load the image again
dir_path = os.path.dirname(os.path.realpath('__file__'))
filename = dir_path + "./images/MarshOrchid.jpg"
raw_image_data = mpimg.imread(filename)

image = tf.placeholder("uint8", [None, None, 3])
# tf.slice(
#    input_,
#    begin,
#    size,
#    name=None
#)
# This operation extracts a slice of size size from a tensor input starting at the location 
# specified by begin. The slice size is represented as a tensor shape, where size[i] is the 
# number of elements of the 'i'th dimension of input that you want to slice. The starting 
# location (begin) for the slice is represented as an offset in each dimension of input. 
# In other words, begin[i] is the offset into the 'i'th dimension of input that you want to slice from.
slice = tf.slice(image, [1000, 0, 0], [3000, -1, -1])

with tf.Session() as session:
    result = session.run(slice, feed_dict={image: raw_image_data})
    print(result.shape)

plt.imshow(result)
plt.show()
```
